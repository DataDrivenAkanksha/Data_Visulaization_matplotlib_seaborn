Understanding plt.rcParams['figure.figsize'] = [6, 3] in Matplotlib
This line sets the default figure size for all Matplotlib plots.

Breakdown:
plt.rcParams → A dictionary that controls Matplotlib settings (e.g., fonts, colors, figure sizes, etc.).
'figure.figsize' → Key that defines the width and height (in inches) of all figures.
[6, 3] → Width = 6 inches, Height = 3 inches

```python
plt.rcParams['figure.figsize']=[6,3]
```

```python
# Pie Chart
values=[4,7,8,9,10,12]
#colors
col=['b','g','c','m','y','r']
# Can use below website to take the hex number of the color code: best practice to define color in this format. 
#https://www.w3schools.com/colors/colors_picker.asp
#label
lab=['A','B','C','D','E','F']
#cut slice (Explode)
exp=[0,0,0,0,0,0.2]
plt.pie(values,colors=col,labels=lab,explode=exp,autopct='%.1f%%',counterclock=False) #autopct--> string formatting for display of percentage
plt.show()
```

```python
# Pie Chart
values=[4,7,8,9,10,12]
#colors
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ffff33','#ff4d4d']
# Can use below website to take the hex number of the color code: best practice to define color in this format. 
#https://www.w3schools.com/colors/colors_picker.asp
#label
lab=['A','B','C','D','E','F']
#cut slice (Explode)
exp=[0,0,0,0,0,0]
plt.pie(values,colors=col,labels=lab,explode=exp,autopct='%.1f%%',counterclock=False) #autopct--> string formatting for display of percentage
plt.show()
```

```python
#pie chart
country=['Germany','USA','India','Russia','Sinngapore']
medal=[20,30,50,25,8]
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ff4d4d']
exp=[0,0,0.1,0,0]
plt.pie(medal,explode=exp,colors=col,labels=country,autopct='%1.2f%%',shadow=True)
#autopct: This parameter is used in the plt.pie() function to format the percentage values on the pie slices.
# '%4.2f%%':
# 4 → The total width of the number (including decimal and percentage sign).
# .2f → Display the percentage with 2 decimal places.
# %% → The double percent (%%) ensures that the percentage sign (%) appears in the output.
plt.title('Medal Tally',fontsize=18)
plt.show()
```

```python
city=['Delhi','Beijing','Washington','Tokyo','Moscow']
happiness_index=[60,40,70,65,85]
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ff4d4d']
plt.xlabel('city')
plt.ylabel('happiness index')
plt.title('Happinss Graph',fontsize=15)
plt.grid()
plt.bar(city,happiness_index,color=col)

# to print the qty of each bar on top of it
for x,y in zip(city,happiness_index):
    plt.text(x,y+2,y)
plt.show()
```

```python
# Grouped bar chart
x=np.arange(5)
y1=[22,4,13,34,25]
y2=[33,44,55,66,77]
width=0.2
# plot data in grouped manner of bar type
plt.bar(x-0.2,y1,width)
plt.bar(x+0.2,y2,width)
for i,j in zip(x,y1):
    plt.text(i-0.3,j+2,j)
    
for i,j in zip(x,y2):
    plt.text(i+0.1,j+2,j)
plt.show()

```
