In the state column, replace the -1 with the mode of state column.

```python
# Ensure that the state and region columns are in the right type (string or numeric)
indian_food_df['state'] = indian_food_df['state'].astype(str)
indian_food_df['region'] = indian_food_df['region'].astype(str)
# Calculate mode
mode_state = indian_food_df['state'].mode()[0]
print("Mode State:", mode_state)
mode_region = indian_food_df[indian_food_df['state'] == mode_state]['region'].mode()[0]
print("Mode Region:", mode_region)


#In the state column, replace the -1 with the mode of state column
indian_food_df['state'] = indian_food_df['state'].replace('-1', mode_state)
#Whatever the state you got in previous question, fill the corresponding region in the region column.
indian_food_df['region'] = indian_food_df['region'].replace('-1', mode_region)

#Replace the -1 with mean for cook time variable.
mean_cook_time=indian_food_df[indian_food_df['cook_time'] >0]['cook_time'].mean()
print("Mean Cook Time:", mean_cook_time)
indian_food_df['cook_time'] = indian_food_df['cook_time'].replace(-1, mean_cook_time)
indian_food_df
```

```python
indian_food_df.to_excel('C:\\Users\\rawat\\Downloads\\python\\data_visualization\\indian_food_df.xlsx')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='cook_time', data=indian_food_df)

# Formatting
plt.xticks(rotation=45)
plt.xlabel('Region')
plt.ylabel('Cook Time')
plt.title('Distribution of Cook Time Across Regions')

plt.show()
```

```python
#d)	Most preferred flavours by number of customers using pie chart.
#pie chart
flavor_counts = indian_food_df['flavor_profile'].value_counts()
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ff4d4d'][:len(indian_food_df['flavor_profile'])]
#exp=[0,0,0,0,0]
plt.pie(flavor_counts, labels=flavor_counts.index, autopct='%1.2f%%', colors=col, explode=None, shadow=True)

plt.title('Flavours by number of customers',fontsize=18)
plt.show()
```

```python
#e)	Most requested course by customers using Bar chart. Print the labels also on the top of each bar. Then sort the bar chart in descending order.

course_count=indian_food_df['course'].value_counts()
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ff4d4d'][:len(indian_food_df['course'])]
plt.xlabel('course')
plt.ylabel('course_order')
plt.title('Most requested course by customers',fontsize=15)
plt.grid()
plt.bar(course_count.index,course_count,color=col)

# to print the qty of each bar on top of it
for x,y in zip(course_count.index,course_count):
    plt.text(x,y+2,y)
plt.show()
```

```python
#f)	Comparison between Vegetarians and Non-Veg orders in each region. Hint: You can use stacked/grouped bar chart for comparison.
import numpy as np
import matplotlib.pyplot as plt

# Get unique regions from the DataFrame
regions = indian_food_df['region'].unique()

# Count vegetarian and non-vegetarian food in each region
veg_count = indian_food_df[indian_food_df['Diet'] == 'vegetarian']['region'].value_counts().reindex(regions, fill_value=0)
non_veg_count = indian_food_df[indian_food_df['Diet'] == 'non vegetarian']['region'].value_counts().reindex(regions, fill_value=0)

# Convert categorical regions to numerical indices
x = np.arange(len(regions))
width = 0.4  

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, veg_count, width, label='Vegetarian', color='green')
plt.bar(x + width/2, non_veg_count, width, label='Non-Vegetarian', color='red')

# Add data labels
for i, j in zip(x, veg_count):
    plt.text(i - (width/2), j + 1, j)  # Adjust the positioning for readability
    
for i, j in zip(x, non_veg_count):
    plt.text(i + (width/2), j + 1, j)  # Adjust the positioning for readability

# Set labels and title
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Comparison of Vegetarian and Non-Vegetarian Dishes by Region')

# Add x-ticks (regions)
plt.xticks(x, regions, rotation=45)

# Add legend
plt.legend()

# Show plot

plt.show()


```

g)	Create the box plot for Order cooking time to see the distribution. Also check if there are any outliers in this column.

```python
cook_time=indian_food_df['cook_time']
lab=['cooking time']
plt.boxplot(cook_time,vert=0,patch_artist=True,labels=lab,showfliers=1)
# vert=0 for horizontal and 1 for vertical
plt.title('Order cooking time distribution using box plot')
plt.show()

# yes we see outliers at 120 and 720.
```

```python
import seaborn as sns

# Create a correlation matrix
corr_matrix = Billionaires_Stats_df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create a heatmap using seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Set the title
plt.title('Correlation Heatmap of Variables')

# SAve and Show the plot
plt.savefig('Heat map.jpg')
plt.show()
```

```python
# Line Chart
import numpy as np
import matplotlib.artist as artist #matplotlib.artist is a module in Matplotlib that provides a base class (Artist) for all objects that are drawn on a figure. This includes elements like lines, text, and patches.
x=range(1,11)
#list(x)
y=[4,6,7,9,5,3,10,9,12,15]
#list(y)
plt.plot(x,y) # this will also generate the plot, but better practice is to use plt.show()
plt.xlim(0,12) #limit range, we are fixing the x axis limit
plt.ylim(0,20) #limit range, we are fixing the y axis limit
plt.show()  # this command to generate the line plot
```

```python
# Pie Chart
values=[4,7,8,9,10,12]
#colors
col=['b','g','c','m','y','r']
# Can use below website to take the hex number of the color code: best practice to define color in this format. 
#https://www.w3schools.com/colors/colors_picker.asp
#label
lab=['A','B','C','D','E','F']
#cut slice (Explode)
exp=[0,0,0,0,0,0.2]
plt.pie(values,colors=col,labels=lab,explode=exp,autopct='%.1f%%',counterclock=False) #autopct--> string formatting for display of percentage
plt.show()
```

```python
# create bar chart
x=range(1,7)
y=[5,6,7,8,9,10]
col=['#4d4dff','#66ff66','#33ffff','#c44dff','#ff4d4d']
plt.bar(x,y,color=col)
plt.savefig('bar chart.jpg') #save fig
plt.title("dummy chart",fontsize=12)
plt.show()
```

```python
# Histogram
values=[80,50,67,78,45,55,89,45,66,56,100,98,65,110]
plt.hist(values,bins=7,color='#66ff68',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
